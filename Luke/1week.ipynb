```python
#01-30 모의고사
def solution(answers):
    # 1. 수포자들의 반복 패턴 정의
    p1 = [1, 2, 3, 4, 5]
    p2 = [2, 1, 2, 3, 2, 4, 2, 5]
    p3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    
    # 2. 각 수포자의 점수를 저장할 리스트
    scores = [0, 0, 0]
    
    # 3. 정답과 비교 (나머지 연산으로 패턴 반복 처리)
    for idx, answer in enumerate(answers):
        if answer == p1[idx % len(p1)]:
            scores[0] += 1
        if answer == p2[idx % len(p2)]:
            scores[1] += 1
        if answer == p3[idx % len(p3)]:
            scores[2] += 1
            
    # 4. 가장 높은 점수 찾기
    max_score = max(scores)
    
    # 5. 최다 득점자 추출 (오름차순)
    result = []
    for i, score in enumerate(scores):
        if score == max_score:
            result.append(i + 1)
            
    return result
```

```python
#02-02 폰켓몬

def solution(nums):
    max_pick = len(nums) // 2
    
    unique_types = len(set(nums))
    
    return min(max_pick, unique_types)
```

```python
#02-03 숫자 문자열과 영단어

def solution(s):
    num_dict = {
        "zero": "0", "one": "1", "two": "2", "three": "3", "four": "4",
        "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9"
    }
    
    answer = s
    for key, value in num_dict.items():
        answer = answer.replace(key, value)
        
    return int(answer)
```

```python
# 02-04 행렬 덧셈

def solution(arr1, arr2):
    rows = len(arr1)
    cols = len(arr1[0])
    
    answer = []
    
    for i in range(rows):
        row_sum = []
        for j in range(cols):
            row_sum.append(arr1[i][j] + arr2[i][j])
        answer.append(row_sum)
        
    return answer
```
